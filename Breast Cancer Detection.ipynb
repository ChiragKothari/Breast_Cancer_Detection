```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib notebook
```

Import breast cancer datast from sklearn dataset

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
#print(cancer.DESCR)
```

Lets first covert this dataset into a dataframe

```python
data = np.vstack((cancer.data.T, cancer.target)).T
columns = np.append(cancer.feature_names, ['target'])
cancerdf = pd.DataFrame(data, columns = columns)
cancerdf.head()
```

```python
X = cancerdf[cancer.feature_names]
y = cancerdf['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

# k-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 8)
knn.fit(X_train, y_train)
print('Training Accuracy = {:.3f}'.format(knn.score(X_train, y_train)))
print('Test Accuracy = {:.3f}'.format(knn.score(X_test, y_test)))
```

Lets visualize this classifier and find the best possible value of n_neighbors that gives the best prediction

```python
neighbors_value = range(1, 11)
training_accuracy = []
test_accuracy = []
for n_neighbors in neighbors_value:
    clf = KNeighborsClassifier(n_neighbors = n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
plt.figure(figsize= (8,4))
plt.plot(neighbors_value, training_accuracy, label = 'Training Accuracy')
plt.plot(neighbors_value, test_accuracy, label = "Test Accuracy")
plt.xlabel('n_neighbors')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

From graph we can see that n_neighbors = 8 will give the best generalization.

# Linear Models for Classification
## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(C = 36).fit(X_train, y_train)
print('Training Accuracy = {:.3f}'.format(logreg.score(X_train, y_train)))
print('Test Accuracy = {:.3f}'.format(logreg.score(X_test, y_test)))
```

Lets visualize this classifier

```python
c_value = range(1, 101)
training_accuracy1 = []
test_accuracy1 = []
for C in c_value:
    logclf = LogisticRegression(C = C).fit(X_train, y_train)
    training_accuracy1.append(logclf.score(X_train, y_train))
    test_accuracy1.append(logclf.score(X_test, y_test))
plt.figure(figsize = (8,4))
plt.plot(c_value, training_accuracy1, label = 'Training Accuracy')
plt.plot(c_value, test_accuracy1, label = 'Test Accuracy')
plt.legend()
plt.show()
```

From figure it can be noted that value of C = 36 will give the best accuracy.

Now lets try with setting penalty = l1 in logistic regression parameter.

```python
logreg_l1 = LogisticRegression(C = 12, penalty = 'l1').fit(X_train, y_train)
print('Training Accuracy = {:.3f}'.format(logreg_l1.score(X_train, y_train)))
print('Test Accuracy = {:.3f}'.format(logreg_l1.score(X_test, y_test)))
```

It can be noted that l1 penalty has increased the test accuracy while training accuracy remained almost same!

```python
training_accuracy_l1 = []
test_accuracy_l1 = []
for C in c_value:
    logclf_l1 = LogisticRegression(C = C, penalty = 'l1').fit(X_train, y_train)
    training_accuracy_l1.append(logclf_l1.score(X_train, y_train))
    test_accuracy_l1.append(logclf_l1.score(X_test, y_test))
plt.figure(figsize = (8,4))
plt.plot(c_value, training_accuracy_l1, label = 'Training Accuracy')
plt.plot(c_value, test_accuracy_l1, label = 'Test Accuracy')
plt.legend()
plt.show()
```

l1 penalty consumed more time.
Well value of C = 12 will be best fit!

## Linear Support Vector Classifier

```python
from sklearn.svm import LinearSVC
linsvc = LinearSVC(random_state = 18).fit(X_train, y_train)
print('Training Accuracy = {:.3f}'.format(linsvc.score(X_train, y_train)))
print('Test Accuracy = {:.3f}'.format(linsvc.score(X_test, y_test)))
```

SVC splits the dataset by own. So, lets see which random_state can provide us best generalization.

```python
random_value = range(1,41)
training_accuracy2 = []
test_accuracy2 = []
for random_state in random_value:
    linsvc = LinearSVC(random_state = random_state).fit(X_train, y_train)
    training_accuracy2.append(linsvc.score(X_train, y_train))
    test_accuracy2.append(linsvc.score(X_test, y_test))
plt.figure(figsize = (8,4))
plt.plot(random_value, training_accuracy2, label = 'Training Accuracy')
plt.plot(random_value, test_accuracy2, label = "Test Accuracy")
plt.xlabel('Random_state values')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

random_state = 18 will be the best fit!

# Neural Networks

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(random_state = 0)
mlp.fit(X_train, y_train)
print('Training Accuracy = {:.3f}'.format(mlp.score(X_train, y_train)))
print('Test Accuracy = {:.3f}'.format(mlp.score(X_test, y_test)))
```

Lets rescale our data for better results.

```python
mean_on_train = X_train.mean(axis = 0)
std_on_train = X_train.std(axis = 0)
X_train_scaled = (X_train - mean_on_train) / std_on_train
X_test_scaled = (X_test - mean_on_train) / std_on_train
mlp = MLPClassifier(max_iter = 1000, alpha = 1, random_state = 0)
mlp.fit(X_train_scaled, y_train)
print('Training Accuracy = {:.3f}'.format(mlp.score(X_train_scaled, y_train)))
print('Test Accuracy = {:.3f}'.format(mlp.score(X_test_scaled, y_test)))
```

Better accuracy rates are achieved through rescaling!

Thus we can predict whether a breast cancer patient is malignant or benign with an accuracy of 97.90%

It is also notable that many of the well-performing models achieved exactly the same highest accuracy of 0.979.
Thus fro this it can be concluded that all of the models make exactly the same number of mistakes.
